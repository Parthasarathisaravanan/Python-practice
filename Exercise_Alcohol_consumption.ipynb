```python
drinks = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv")
drinks
```

### Step 4. Which continent drinks more beer on average?

```python
res=drinks.groupby(['continent']).sum()
res.sum(axis=1).head(1)
```

### Step 5. For each continent print the statistics for wine consumption.

```python
res['wine_servings']
```

### Step 6. Print the mean alcoohol consumption per continent for every column

```python
#drinks.merge(res,on='continent')
#pd.merge(drinks, res, on='continent')
res['mean_alcohol'] = res.mean(axis=1)
#res=res.drop(['beer_servings', 'spirit_servings','wine_servings','total_litres_of_pure_alcohol'], axis = 1)
res
```

```python
#res=res.drop(['beer_servings', 'spirit_servings','wine_servings','total_litres_of_pure_alcohol'], axis = 1)
ans=pd.merge(drinks,res,on='continent')
ans
```

### Step 7. Print the median alcoohol consumption per continent for every column

```python
res1=drinks.groupby(['continent']).sum()
res1.sum(axis=1).head(1)
res['median_alcohol']=res1.median(axis=1)
res
```
